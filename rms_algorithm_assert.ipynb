```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('test_out.csv')

iterations = data[data.columns[0]]
adc_value = data[data.columns[1]]
real_rms_value = data[data.columns[2]]
calculated_rms_value_2 = data[data.columns[3]]
error_per_2 = (calculated_rms_value_2 - real_rms_value)/real_rms_value

## Plot Results
alpha=0.7
f, (a0, a1) = plt.subplots(2, 1, figsize=(12,6), gridspec_kw={'height_ratios': [5, 1]})
a0.plot(iterations, real_rms_value, label='Real RMS Value', alpha=alpha)
a0.plot(iterations, calculated_rms_value_2, '--', label='Calculated RMS Value', alpha=alpha)
a0.legend()
a0.set_ylim((150, 300))
a0.set_title('RMS Algorithm Test')
a1.plot(iterations, error_per_2, label='Calculated RMS Value', alpha=alpha)
a1.legend()
a1.set_ylim((-0.05, 0.05))
a1.set_title('Percentage Error')

plt.tight_layout()
plt.show()
plt.savefig('real_time_rms_algorithm_test_in_c.png')
```
